# Regressão Linear Múltipla
Dataset: Carsmall

## Importação de Bibliotecas

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## Importação do Dataset

```
df = pd.read_csv('/content/drive/My Drive/PORTIFOLIO/Algoritmos Machine Learning/Datasets/carsmall.csv')
df.head()
```

```
df.tail()
```

## Treinamento do Modelo

```
# Atribuindo valores das colunas Quilates e Preco para as variáveus x e y

X = df[['Weight','Model_Year_76', 'Model_Year_82']] #independentes
y = df['MPG'] #dependentes

# Plotar gráfico de disperção das variáveis x e y

modelo1 = LinearRegression()
modelo1.fit(X,y)
```

```
# Exibe os valores do coeficiente linear e angular do modelo1
print('Coeficiente linear é:', modelo1.intercept_)
print('Coeficiente angulares são (beta1,beta2,beta3):', modelo1.coef_)

# Exibe o valor de R2 do modelo1
print('Coeficiente R2 é: ', modelo1.score(X,y))
```

## Transcrição do Modelo
Modelo: Consumo em MPG = 39,897 - 6,584.Peso + 1,95.Ano_76 + 8,13.Ano_82

## Representação Gráfica


```
X_70 = df[df['Model_Year_70'] == 1]
X_76 = df[df['Model_Year_76'] == 1]
X_82 = df[df['Model_Year_82'] == 1]
 
plt.scatter(X_70['Weight'],X_70['MPG'], color='red')
plt.scatter(X_76['Weight'],X_76['MPG'], color='blue')
plt.scatter(X_82['Weight'],X_82['MPG'], color='green')
plt.title("Diagrama de dispersão: y = f(x)")
plt.xlabel("Peso")
plt.ylabel("Consumo")
plt.show()
```
