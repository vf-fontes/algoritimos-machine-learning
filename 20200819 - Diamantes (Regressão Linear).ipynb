# Regressão Linear
dataset: Diamantes

## Importação das Biliotecas

```
# Importações de bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## Importação do Dataset

```
# Leitura do Dataset
df = pd.read_csv('/content/drive/My Drive/PORTIFOLIO/Algoritmos Machine Learning/Datasets/diamantes.csv')
df.head()
```

## Verificando graficamente se as variáveis "preço" e "peso" aparentam ter um relacionamento linear positivo.

```
# Atribuindo valores das colunas Quilates e Preco para as variáveus x e y

x = df.Quilates.values.reshape(-1,1)
y = df.Preco.values

# Plotar gráfico de disperção das variáveis x e y

plt.scatter(x,y)
plt.title('Relação entre Quilates e Preço')
plt.xlabel('Quilate')
plt.xlabel('Preço')
```

## Treinamento do Modelo de Regressão Linear


```
# Instancia a regressão linear
modelo1 = LinearRegression()
modelo1.fit(x,y)
```

## Previsão do Modelo

```
# Exibe os valores do coeficiente linear e angular do modelo1
print("Coeficiente linear =",modelo1.intercept_)
print("Coeficientes angulares =",modelo1.coef_)
```

## Métrica R2

```
# Exibe o valor de R2 do modelo1
print(modelo1.score(x,y))
```

## Testando o Modelo para um diamante de 0,45 quilate


```
# Realiza um teste
x_teste = np.array([0.45]).reshape(-1,1)
print('Preço justo para o diamante de 0.45 quilates é:', modelo1.predict(x_teste))
```
