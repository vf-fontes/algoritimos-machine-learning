# Regressão Linear
dataset: defeitos

## Importação das Bibliotecas

```
# Realizando importações

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## Importação do Dataset

```
# Importar Dataset
df = pd.read_csv('/content/drive/My Drive/PORTIFOLIO/Algoritmos Machine Learning/Datasets/defeitos.csv')
df.head()
```

## Selecionando variáveis para o treinamento do Modelo

```
# Atribuindo valores das colunas Quilates e Preco para as variáveus x e y

X = df[['x1','x2']]
y = df['y']
```

## Treinando o Modelo

```
modelo1 = LinearRegression()
modelo1.fit(X,y)
```

## Previsão do Modelo

```
# Exibe os valores do coeficiente linear e angular do modelo1
print('Coeficiente linear é:', modelo1.intercept_)
print('Coeficiente angulares são (beta1,beta2):', modelo1.coef_)
```

## Métrica R2

```
# Exibe o valor de R2 do modelo1
print('Coeficiente R2 é: ', modelo1.score(X,y))
```

## Testando o Modelo para uma máquina produza 30 unidades/hora e esteja a duas semanas sem passar por manutenção. A contagem muda caso a máquina fique sem manutenção por mais uma semana?


```
# Realiza um teste
x_teste = np.array([[30,2],[30,3]]).reshape(-1,2)
print('Defeitos para (x1,x2) = (30,2) e (30,3):', modelo1.predict(x_teste))
```

Sim o valor muda, porém não significativamente a ponto de ser 1 peça defeituosa a mais.
